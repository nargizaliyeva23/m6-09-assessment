1. **Missed detections in challenging lighting conditions**

   - Some images resulted in low-confidence predictions or no detections at all.

2. **Difficulty generalising across different cat appearances**
   - The model sometimes struggled to detect cats with unusual poses, fur colours, or breeds that appeared less frequently in the training dataset.
   - Detection performance decreased when cats looked significantly different from the majority of training examples.

## ONNX Validation

The exported ONNX model was loaded using ONNX Runtime and tested on sample images to verify that the exported model produces predictions consistent with the original PyTorch model.

````python
import onnxruntime as ort

session = ort.InferenceSession(
    "C:\\Users\\Asus\\Documents\\IRONHACK\\Labs\\m6-09-assessment\\latest_best.onnx",
    providers=["CPUExecutionProvider"]
)

print("Input shape:", session.get_inputs()[0].shape)
print("Output shape:", session.get_outputs()[0].shape)
````

The exported ONNX model was successfully loaded and validated using ONNX Runtime.

### Model Input and Output Shapes

- **Input Shape:** `(1, 3, 640, 640)`
- **Output Shape:** `(1, 300,6)`

The input tensor represents:

- Batch size = 1
- RGB image with 3 channels
- Resolution = 640 × 640 pixels

The output tensor follows the YOLO26 end-to-end detection format:

`[x1, y1, x2, y2, confidence, class]`

where:

- `x1, y1` = top-left corner of the bounding box
- `x2, y2` = bottom-right corner of the bounding box
- `confidence` = detection confidence score
- `class` = predicted object class

The model can return up to 300 detections per image.

The successful validation confirms that the ONNX export was completed correctly and that the model is ready for deployment and inference using ONNX Runtime.

````python
#PyTorch
from ultralytics import YOLO

pt_model = YOLO("C:/Users/Asus/Documents/IRONHACK/Labs/m6-09-assessment/latest_best.pt")

pt_results = pt_model(
    "C:/Users/Asus/Downloads/cat_task.jpg",
    conf=0.25
)

print(pt_results[0].boxes.xyxy)
````

````python
#ONNX
onnx_model = YOLO("C:/Users/Asus/Documents/IRONHACK/Labs/m6-09-assessment/latest_best.onnx")

onnx_results = onnx_model(
    "C:/Users/Asus/Downloads/cat_task.jpg",
    conf=0.25
)

print(onnx_results[0].boxes.xyxy)
````

````python
pt_boxes = pt_results[0].boxes.xyxy.cpu().numpy()
onnx_boxes = onnx_results[0].boxes.xyxy.cpu().numpy()

print("PyTorch detections:", len(pt_boxes))
print("ONNX detections:", len(onnx_boxes))
````
